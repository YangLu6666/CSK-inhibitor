```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import seaborn as sns

# 读取 CSV 文件
data = pd.read_csv(r'D:\Postdoctor\VS\CSK\CSK-CHEMBL.csv', encoding='ISO-8859-1')

# 转换 SMILES 为分子指纹
def smiles_to_fingerprint(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        fingerprint = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
        return np.array(fingerprint)
    else:
        return None

# 应用函数并去除无效 SMILES
data['Fingerprint'] = data['Smiles'].apply(smiles_to_fingerprint)
data = data[data['Fingerprint'].notnull()]

# 创建指纹矩阵
fingerprint_matrix = np.array(data['Fingerprint'].tolist())

# 使用 t-SNE 进行降维
tsne = TSNE(n_components=2, random_state=42)
tsne_results = tsne.fit_transform(fingerprint_matrix)

# 将 t-SNE 结果添加到数据框中
data['TSNE-1'] = tsne_results[:, 0]
data['TSNE-2'] = tsne_results[:, 1]

# 绘制 t-SNE 图
plt.figure(figsize=(10, 8))
sns.scatterplot(x='TSNE-1', y='TSNE-2', hue='Standard Value', data=data, palette='viridis', alpha=0.7)
plt.title('t-SNE Clustering of Compounds')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend(title='Standard Value')
plt.grid()

# 保存为 SVG 格式
plt.savefig(r'D:\Postdoctor\VS\CSK\tsne_clustering.svg', format='svg')
plt.show()
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem, MACCSkeys
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 设置图形风格
sns.set(style="whitegrid")

# 读取 CSV 文件
data = pd.read_csv(r'D:\Postdoctor\VS\CSK\CSK-CHEMBL.csv', encoding='ISO-8859-1')

# 转换 SMILES 为分子指纹
def smiles_to_fingerprints(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        morgan_fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
        maccs_fp = MACCSkeys.GenMACCSKeys(mol)
        return np.concatenate([np.array(morgan_fp), np.array(maccs_fp)])
    else:
        return None

# 应用函数并去除无效 SMILES
data['Fingerprint'] = data['Smiles'].apply(smiles_to_fingerprints)
data = data[data['Fingerprint'].notnull()]

# 创建指纹矩阵和标签
fingerprint_matrix = np.array(data['Fingerprint'].tolist())
labels = (data['Standard Value'] > 5).astype(int)  # 以 5 为界限进行分类

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(fingerprint_matrix, labels, test_size=0.2, random_state=42)

# 定义模型
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Support Vector Machine': SVC(probability=True, random_state=42)
}

# 绘制 AUC 曲线
plt.figure(figsize=(10, 8))

for model_name, model in models.items():
    # 训练模型
    model.fit(X_train, y_train)
    
    # 预测概率
    y_scores = model.predict_proba(X_test)[:, 1]
    
    # 计算 ROC 曲线
    fpr, tpr, _ = roc_curve(y_test, y_scores)
    roc_auc = auc(fpr, tpr)
    
    # 绘制 AUC 曲线
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')

# 绘制随机猜测的基线
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess (AUC = 0.5)')

# 图形设置
plt.title('ROC Curves of Different Models')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()

# 保存为矢量图格式（SVG）
plt.savefig(r'D:\Postdoctor\VS\CSK\auc_curves.svg', format='svg')
plt.show()
```
